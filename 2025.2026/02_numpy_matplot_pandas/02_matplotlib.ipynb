# Kratak pregled `matplotlib` biblioteke

U ovoj svesci prikazujemo osnovne tipove grafika koje cemo koristiti kasnije na kursu.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Linijski grafik

```python
# Uzimamo 100 ekvidistantnih tacaka iz intervala [1, 5]
x = np.linspace(1, 5, 100)

# Definisemo funkcije
y0 = x
y1 = x * np.log(x)
y2 = np.sqrt(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y0, label='x', linewidth=2)
plt.plot(x, y1, label='x * log(x)', linewidth=2)
plt.plot(x, y2, label='sqrt(x)', linewidth=2)
plt.title('Poredjenje funkcija')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

## Scatter grafik

```python
rng = np.random.default_rng(7)
x = rng.normal(0, 1, 80)
y = 0.7 * x + rng.normal(0, 0.6, 80)

plt.figure(figsize=(6, 4))
plt.scatter(x, y, alpha=0.7, edgecolors='black')
plt.title('Scatter primer')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, alpha=0.3)
plt.show()
```

## Histogram

```python
samples = rng.normal(loc=170, scale=8, size=300)

plt.figure(figsize=(6, 4))
plt.hist(samples, bins=20, color='tab:blue', edgecolor='black', alpha=0.8)
plt.title('Histogram nasumicnog uzorka')
plt.xlabel('vrednost')
plt.ylabel('frekvencija')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

## Subplot i stilizacija

```python
x = np.linspace(0, 2 * np.pi, 200)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].plot(x, np.sin(x), color='tab:green')
axes[0].set_title('sin(x)')
axes[0].grid(True, alpha=0.3)

axes[1].plot(x, np.cos(x), color='tab:red')
axes[1].set_title('cos(x)')
axes[1].grid(True, alpha=0.3)

fig.suptitle('Primer subplot prikaza')
plt.tight_layout()
plt.show()
```
