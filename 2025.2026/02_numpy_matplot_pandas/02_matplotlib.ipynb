# Kratak pregled `matplotlib` biblioteke

U ovoj svesci prikazujemo osnovne tipove grafika koje cemo koristiti kasnije na kursu.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Linijski grafik

```python
# Uzimamo 100 ekvidistantnih tacaka iz intervala [1, 5]
x = np.linspace(1, 5, 100)

# Definisemo funkcije
y0 = x
y1 = x * np.log(x)
y2 = np.sqrt(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y0, label='x', linewidth=2)
plt.plot(x, y1, label='x * log(x)', linewidth=2)
plt.plot(x, y2, label='sqrt(x)', linewidth=2)
plt.title('Poredjenje funkcija')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

## Scatter grafik

```python
rng = np.random.default_rng(7)
x = rng.normal(0, 1, 80)
y = 0.7 * x + rng.normal(0, 0.6, 80)

plt.figure(figsize=(6, 4))
plt.scatter(x, y, alpha=0.7, edgecolors='black')
plt.title('Scatter primer')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, alpha=0.3)
plt.show()
```

## Histogram

```python
samples = rng.normal(loc=170, scale=8, size=300)

plt.figure(figsize=(6, 4))
plt.hist(samples, bins=20, color='tab:blue', edgecolor='black', alpha=0.8)
plt.title('Histogram nasumicnog uzorka')
plt.xlabel('vrednost')
plt.ylabel('frekvencija')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

## Subplot i stilizacija

```python
x = np.linspace(0, 2 * np.pi, 200)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].plot(x, np.sin(x), color='tab:green')
axes[0].set_title('sin(x)')
axes[0].grid(True, alpha=0.3)

axes[1].plot(x, np.cos(x), color='tab:red')
axes[1].set_title('cos(x)')
axes[1].grid(True, alpha=0.3)

fig.suptitle('Primer subplot prikaza')
plt.tight_layout()
plt.show()
```

## Dodatni referentni primeri `Matplotlib` funkcija

U nastavku su dati kratki primeri za najčešće korišćene `matplotlib.pyplot` funkcije. Za svaku funkciju prikazana su barem dva tipična poziva.

```python
x = np.linspace(0, 2 * np.pi, 200)
y_sin = np.sin(x)
y_cos = np.cos(x)
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 6]
image = np.arange(16).reshape(4, 4)
```

```python
# plt.style.use, plt.figure, plt.plot, plt.title, plt.xlabel, plt.ylabel, plt.legend, plt.grid, plt.show
plt.style.use('classic')
plt.figure(figsize=(7, 3))
plt.plot(x, y_sin, label='sin(x)', linewidth=2)
plt.plot(x, y_cos, label='cos(x)', linewidth=2)
plt.title('Linijski grafik - stil classic')
plt.xlabel('x')
plt.ylabel('vrednost')
plt.legend()
plt.grid(True)
plt.show()

plt.style.use('default')
plt.figure(figsize=(7, 3))
plt.plot(x, x / np.pi, label='x / pi', color='tab:purple')
plt.plot(x, np.sqrt(x), label='sqrt(x)', color='tab:green')
plt.title('Linijski grafik - stil default')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# plt.scatter, plt.xlim, plt.ylim, plt.xticks, plt.yticks, plt.show
rng = np.random.default_rng(0)
x_scatter = rng.normal(size=60)
y_scatter = 0.8 * x_scatter + rng.normal(scale=0.4, size=60)

plt.figure(figsize=(6, 4))
plt.scatter(x_scatter, y_scatter, alpha=0.7, c='tab:blue')
plt.title('Scatter primer 1')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.xticks([-3, -2, -1, 0, 1, 2, 3])
plt.yticks([-3, -2, -1, 0, 1, 2, 3])
plt.grid(alpha=0.3)
plt.show()

plt.figure(figsize=(6, 4))
plt.scatter(x_scatter, y_scatter, c=x_scatter, cmap='viridis', edgecolors='black')
plt.title('Scatter primer 2')
plt.xlim(-2.5, 2.5)
plt.ylim(-2.5, 2.5)
plt.xticks([-2, -1, 0, 1, 2])
plt.yticks([-2, -1, 0, 1, 2])
plt.grid(alpha=0.2)
plt.show()
```

```python
# plt.bar, plt.hist, plt.tight_layout, plt.show
plt.figure(figsize=(6, 3))
plt.bar(categories, values, color='tab:orange')
plt.title('Bar primer 1')
plt.xlabel('kategorija')
plt.ylabel('vrednost')
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 3))
plt.bar(['P', 'Q', 'R'], [2, 4, 3], color=['tab:red', 'tab:green', 'tab:blue'])
plt.title('Bar primer 2')
plt.tight_layout()
plt.show()

samples = rng.normal(loc=170, scale=8, size=300)
plt.figure(figsize=(6, 3))
plt.hist(samples, bins=15, color='tab:blue', edgecolor='black')
plt.title('Histogram primer 1')
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 3))
plt.hist(samples, bins=20, density=True, color='tab:green', alpha=0.7)
plt.title('Histogram primer 2')
plt.tight_layout()
plt.show()
```

```python
# plt.subplots, plt.imshow, plt.savefig, plt.show
fig, axes = plt.subplots(1, 2, figsize=(8, 3))
axes[0].imshow(image, cmap='viridis')
axes[0].set_title('imshow primer 1')
axes[1].imshow(image, cmap='gray')
axes[1].set_title('imshow primer 2')
plt.tight_layout()
plt.savefig('matplotlib_imshow_primer.png', dpi=120)
plt.show()

fig, axes = plt.subplots(2, 1, figsize=(6, 5))
axes[0].plot(x, y_sin, color='tab:red')
axes[0].set_title('Subplot primer 1')
axes[1].plot(x, y_cos, color='tab:blue')
axes[1].set_title('Subplot primer 2')
plt.tight_layout()
plt.savefig('matplotlib_subplot_primer.png', dpi=120)
plt.show()
```
